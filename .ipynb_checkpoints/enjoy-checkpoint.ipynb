```python
import json
from pathlib import Path
from utils import plot_sample, clean_borders, inp2img
import matplotlib.pyplot as plt
import numpy as np
from models import CAModel, CAModelConvLSTM, MetaConvLSTM_CA
from train import tensorize_concatenate
import torch
from data_loader import DatasetARC

from tqdm import tqdm
from matplotlib import colors
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
model = MetaConvLSTM_CA(11).to(device)
load_model = '/workspace7/Unity3D/gabriele/arc_challenge/RUNS/exp_8/models/arc.state_dict.800'
model.load_state_dict(torch.load(load_model, map_location=device))
tasks = DatasetARC('data/abstraction-and-reasoning-challenge')

# task = tasks[4]['train']
# X = torch.cat([tensorize_concatenate(demo, device) for demo in task['train']], dim=0)
# test_x = torch.from_numpy(inp2img(task['test'][0]['input'])).unsqueeze(0).float().to(device)
# test_y = torch.from_numpy(task['test'][0]['output']).unsqueeze(0).long().to(device)
# y_pred = model(X, test_x, 1)
#y_pred.numpy

cmap = colors.ListedColormap(
        ['#000000', '#0074D9','#FF4136','#2ECC40','#FFDC00',
         '#AAAAAA', '#F012BE', '#FF851B', '#7FDBFF', '#870C25', '#3d9970'])
norm = colors.Normalize(vmin=0, vmax=10)

def animate_solution(model, task):
    #x = torch.from_numpy(inp2img(sample["input"])).unsqueeze(0).float().to(device)
    X = torch.cat([tensorize_concatenate(demo, device) for demo in task['train']], dim=0)
    x = torch.from_numpy(inp2img(task['test'][0]['input'])).unsqueeze(0).float().to(device)
    import ipdb; ipdb.set_trace()
    test_y = torch.from_numpy(task['test'][0]['output']).unsqueeze(0).long().to(device)
    x = model(X, x, 1)
    @torch.no_grad()
    def animate(i):
        pred = model(X, x, i)
        im.set_data(pred.argmax(1).squeeze().cpu().numpy())

    fig, ax = plt.subplots()
    im = ax.imshow(x.argmax(1).squeeze().cpu().numpy(), cmap=cmap, norm=norm)
    return animation.FuncAnimation(fig, animate, frames=20, interval=120)
anim =  animate_solution(model, tasks[99]) 
HTML(anim.to_jshtml())
```
